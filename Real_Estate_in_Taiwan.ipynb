```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
file_path = "Real estate valuation data set.xlsx"
df = pd.read_excel(file_path)

# Clean column names (optional)
df.columns = df.columns.str.strip()

# Rename columns for better readability
df = df.rename(columns={
    'X1 transaction date': 'Transaction Date',
    'X2 house age': 'House Age',
    'X3 distance to the nearest MRT station': 'MRT Distance',
    'X4 number of convenience stores': 'Convenience Stores',
    'X5 latitude': 'Latitude',
    'X6 longitude': 'Longitude',
    'Y house price of unit area': 'Price per Unit Area'
})
```

```python
# 1. Identify the Most Expensive Properties Based on Latitude and Longitude
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['Longitude'], df['Latitude'], c=df['Price per Unit Area'], cmap='YlOrRd', s=100, alpha=0.7)
plt.colorbar(scatter, label='Price per Unit Area')
plt.title('Geographical Distribution of Real Estate Prices in Taipei')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid()
plt.show()
```

The scatter plot on the account of Geographical Distribution provides a thorough visualization of variation in property values across Taipei. The latitude and longitude axes of the chart, combined with variation of color standing for price per unit area, point to some key patterns that would be of vital importance to an analyst manager.

•	The darker red areas, mostly concentrated between about 121.52 and 121.54 longitude and 24.96 and 24.98 latitude, probably represent the central business districts or highly desirable residential zones of Taipei. It is in these areas that high real estate prices manifest strong market demand, signifying premium locations for investment. The lighter yellow points sprinkled toward the periphery tend to indicate lower property values, and thus may point to regions considered less developed or less desirable.

•	This visual distinction is important to the identification of lucrative investment opportunities and the dynamics of urban development. It simply means that areas with high real estate prices are located close to economic hubs, amenities, and transport networks, driving up demand. Lower-priced regions could present development opportunities with future growth potential, especially if infrastructural improvements are planned.

•	This particular analysis underlines the importance of geographic location in property valuation within Taipei. It enables strategic decision-making concerning investments, urban planning, and resource allocation by highlighting areas of high demand and potential growth. Understanding these geographical price disparities is vital for optimizing portfolio performance and capitalizing on market trends.

```python
# 3. Analyze Price Trends Based on MRT Proximity
# Divide MRT distance into bins for easier analysis
df['MRT Distance Category'] = pd.cut(df['MRT Distance'], bins=[0, 500, 1000, 2000, 5000, np.inf],
                                     labels=['0-500m', '500-1000m', '1-2km', '2-5km', '5km+'])

# Calculate the average price per unit area for each distance category
mrt_analysis = df.groupby('MRT Distance Category')['Price per Unit Area'].mean().reset_index()

# Bar plot showing how prices change based on distance to MRT
plt.figure(figsize=(8, 5))
sns.barplot(data=mrt_analysis, x='MRT Distance Category', y='Price per Unit Area', palette='viridis')
plt.title('Average Price per Unit Area Based on MRT Proximity')
plt.xlabel('MRT Distance Category')
plt.ylabel('Average Price per Unit Area')
plt.show()
```

The following bar chart, "Average Price per Unit Area Based on MRT Proximity," reveals some basic characteristics of the Taipei real estate market and the importance and impact of proximity to MRT stations.

### Importance of MRT Proximity

1. **Convenience and Accessibility**: The chart demonstrates that properties closer to MRT stations (0-500m) command the highest prices, surpassing 40 units per area. This underscores the premium placed on convenient access to public transportation, which significantly reduces commuting time and enhances the quality of urban life.

2. **Demand and Supply Dynamics**: Decreasing average prices as the distance from MRT stations increases, with properties 5km+ away having the lowest prices at around 20 units, are indicative of demand-supply dynamics in the real estate market. Properties in close proximity to MRT stations are in high demand, thus fetching higher prices.

### Significance in the Real Estate Market

1. **Urban Development**: The relationship between MRT proximity and property prices provides guidance for strategies of urban planning and development. Potential candidates for high-density residential and commercial projects are the areas close to MRT stations, given the efficiency in land use and the demand in respect of convenient living space.

2. Investment Opportunities: Such dynamics are very important for an investor to understand; not only do properties located near MRT stations command higher prices, but they also assure better appreciation over time, hence more investment opportunities with possibly higher returns. 3. Socio-Economic Implications: The difference in property prices due to the proximity of MRT stations has very important socio-economic implications. Higher prices around those public transportation hubs may create affordability issues for lower-income earners, which could further widen the gap socially.
```python
import folium
import pandas as pd

# Load the dataset
file_path = "Real estate valuation data set.xlsx"
df = pd.read_excel(file_path)

# Clean column names
df.columns = df.columns.str.strip()

# Rename columns for better readability
df = df.rename(columns={
    'X1 transaction date': 'Transaction Date',
    'X2 house age': 'House Age',
    'X3 distance to the nearest MRT station': 'MRT Distance',
    'X4 number of convenience stores': 'Convenience Stores',
    'X5 latitude': 'Latitude',
    'X6 longitude': 'Longitude',
    'Y house price of unit area': 'Price per Unit Area'
})

# Create a base map centered around Taipei (latitude: 25.0330, longitude: 121.5654)
mrt_map = folium.Map(location=[25.0330, 121.5654], zoom_start=12)

# 1. Plot house prices in relation to MRT distance
for idx, row in df.iterrows():
    price = row['Price per Unit Area']
    color = 'green' if price < 20 else 'orange' if price < 40 else 'red'  # Color based on price

    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=7,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.6,
        popup=f"Price: {price}, MRT Distance: {row['MRT Distance']}m"
    ).add_to(mrt_map)

# 2. Mark MRT stations (Assuming sample coordinates for MRT stations in Taipei)
# Note: Replace the coordinates with actual MRT station locations
mrt_stations = [
    {'name': 'Taipei Main Station', 'latitude': 25.0478, 'longitude': 121.5170},
    {'name': 'Zhongxiao Fuxing Station', 'latitude': 25.0399, 'longitude': 121.5250},
    {'name': 'Ximen Station', 'latitude': 25.0320, 'longitude': 121.5075},
]

# Add MRT stations to the map
for station in mrt_stations:
    folium.Marker(
        location=[station['latitude'], station['longitude']],
        popup=station['name'],
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(mrt_map)

# 3. Mark Convenience Stores (Assuming sample coordinates for convenience stores)
# You should replace these with actual convenience store locations in your dataset
for idx, row in df.iterrows():
    if row['Convenience Stores'] > 0:
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"Convenience Stores: {row['Convenience Stores']}",
            icon=folium.Icon(color='green', icon='info-sign')
        ).add_to(mrt_map)

# Save the map to an HTML file
mrt_map.save("House_Prices_MRT_Convenience_Map.html")

# Display the map in Jupyter (if running in Jupyter notebook environment)
mrt_map
```
